##Nested List Comprehension

```
word_1 = "Hello"
word_2 = "World"

result = [i+j for i in word_1 for j in word_2]
print(result)
```

```
word_1 = "Hello"
word_2 = "World"

result = [i+j for i in word_1 for j in word_2 if not (i==j)] # 필터추가
print(result)
```

```
word_1 = "Hello"
word_2 = "World"

result = [i+j if not (i==j) else "NotMach" for i in word_1 for j in word_2] # 필터추가
print(result)
```

#Generator

대용량 데이터 핸들링할 때 자주 쓴다. 함수가 호출되는 시점에 메모리에 할당하기 때문에

메모리 용량을 효율적으로 사용

```
def general_list(value):
    result = []
    for i in range(value):
        result.append(i)
    return result
```

```
print(general_list(20))
```

```
import sys
result = general_list(20)
sys.getsizeof(result) # 해당 함수를 실행하는데 264 바이트 메모리를 차지
```

```
def generator_list(value):
    result = []
    for i in range(value):
        yield i
```

```
result = generator_list(20) # 이 시점에서는 value가 20인 상태만 가지고 있음.
sys.getsizeof(result) # 해당 함수를 실행하는데 88 바이트 메모리를 차지
```

```
for a in generator_list(20):
    print(a)

#프린트 문에서 a를 호출할 때, generator_list에서 yield 로 값을 던져줌
#즉 값이 출력되기 위해 필요로 하는 시점인 그 때, 메모리를 할당하기 때문에. 절약됨
```

```
#예제 1
def asterisk_test(a, args):
    print(a, args)
    print(a, *args)
test = (2,3,4,5,6)
asterisk_test(1, test)
```

```
#예제 2
def asterisk_test(a, *args): # 알규먼트에서 *는 다수의 가변 인자를 받아서 패킹을 해버림
    print(a, args)
    print(a, *args) # 언패킹이 일어남
test = (2,3,4,5,6)
asterisk_test(1, test)
```

```
#예제 3  --> 예제 1과 같은 결과
def asterisk_test(a, *args): # 여기 *은 다수개의 인자를 받을 수 있는 가변인자의 의미
    print(a, args)
    print(a, *args)
test = (2,3,4,5,6)
asterisk_test(1, *test)  # 여기의 *의미는 언패킹 의미
```
