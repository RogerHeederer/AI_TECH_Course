# + 연산자를 붙이는 위치에 따라 달라지는 결과

#shallow copy / deep copy

```
a = [1,2,3]
b = a[:]
id(a), id(b)
```

```
a is b, a == b
```

```
# mutable 객체 안에 mutable 객체는 얕은 복사가 문제가 된다
a = [[1,2],[3,4]]
b = a[:]
id(a), id(b)
```

```
id(a[0]), id(b[0]) # a와 b의 주소는 다른데, 내부에 있는 [1,2]   [3,4]는 같은 곳을 가르키고 있다.
```

```
#재할당은 문제 없음
a[0] = [8,9]
a
```

```
b
```

```
a[1].append(5)
a,b
```

```
#깊은 복사

import copy
a = [[1,2],[3,4]]
b = copy.deepcopy(a)
```

```
# deep 카피하면 위의 얕은 복사와 달리 새로운 주소를 가지게 된다.
id(a[0]), id(b[0])
```

```
a[1].append(5)
```

```
b
```
