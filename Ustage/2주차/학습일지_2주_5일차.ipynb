파이썬 기본 매개변수 유의점

```
def buggy(arg, result=[]):
    result.append(arg)
    print(result)
    
buggy('a') # a
buggy('b') # a, b  a가 누적된다.
```

```
def works(arg):
    result = []
    result.append(arg)
    print(result)

works('a')
works('b')
```

```
def nonbuggy(arg, result=None):
    if result is None:
        result = []
    result.append(arg)
    print(result)

nonbuggy('a')
nonbuggy('b')
```

#위치 인수와 키워드 인수

위치 인수 분해하기/ 모으기

```

def print_args(*args):
    print('Positional tuple:', args)
```

```
print_args()
```

```
print_args(3,2,1,'wait!','uh...')
```

```
args = (2,5,6,'x')
print_args(args)
```

```
# 함수로 전달한 값을 인수라 부름.  함수와 인수를 호출하면, 인수의 값은 함수 내에서 해당되는 매개변수에 복사됨
# 함수 외부에서는 인수. 함수 내부에서는 매개변수
print_args(*args) # 함수 인수로 *args는 언패킹되고, 매개 변수로 받는 *args에서 다시 패킹함
```

키워드 인수 분해하기 / 모으기

```
def print_kwargs(**kwargs):
    print('Keyword arguments:', kwargs)
```

```
print_kwargs()
```

```
print_kwargs(wine='merlot', entree='mutton', dessert='macaroon')
```

```
#키워드 전용 인수

def print_data(data, *, start=2, end=100):
    print(data)
    for value in (data[start:end]):
        print(value)
```

```
data = ['a','b','c','d','e','f']
print(data)
```

```
print_data(data)
```

```
print_data(data, start=4)
```

```
print_data(data, end=4)
```

# 내부 함수

```
def knights(saying):
    def inner(quote):
        return "We are the knights who say: '%s'" % quote
    return inner(saying)
```

```
knights("hi")
```

#클로져

외부 함수에 의해 동적으로 생성되고, 그 외부 함수의 변수값을 알고 있는 함수가 클로져

```
def knights2(saying):
    def inner2():
        return "We are the knights who say: '%s'" % saying # 외부 함수의 매개변수인 saying을 알고 있음
    return inner2
```

```
a = knights2('Duck')
```

```
a # knights2.<locals>.inner2
```

```
type(a)
```

```
a()
```
